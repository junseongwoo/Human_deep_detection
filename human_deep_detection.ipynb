```python
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace
```

## Image 1

```python
img =  cv2.imread("happy1.jpg") 

plt.imshow(img)   
```

```python
color_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow( color_img)
```

```python
prediction = DeepFace.analyze(img)
```

```python
prediction
```

```python
prediction['dominant_emotion']
```

```python
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades +'haarcascade_frontalface_default.xml')
```

```python
faces = faceCascade.detectMultiScale(color_img, 1.1, 4)

for (x, y, u, v) in faces:
    cv2.rectangle(color_img, (x,y), (x+u, y+v), (0, 0, 225), 2)
    
plt.imshow(color_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX

cv2.putText(color_img, prediction['dominant_emotion'], (0, 50), font, 1, 
(225,0,0), 2, cv2.LINE_4)
cv2.putText(color_img, prediction['dominant_race'], (100, 50), font,  1,
(255,0,0), 2, cv2.LINE_4)
cv2.putText(color_img, prediction['gender'], (0, 100), font,  1,
(0,0,0), 2, cv2.LINE_4)
cv2.putText(color_img, str(prediction['age']), (150, 100), font,  1,
(0,0,0), 2, cv2.LINE_4)

plt.imshow(color_img)
```
